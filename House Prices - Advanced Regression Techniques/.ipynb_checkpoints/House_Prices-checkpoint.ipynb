```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Print metrics
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix

def print_metrics(name, y_test, y_pred):
    print('--> ',name, '\n')
    print('Accuracy Score: %.4f'%accuracy_score(y_test,y_pred))
    print('\n Confusion Matix: \n', confusion_matrix(y_test, y_pred))
    print('\n \n Classification Report: ', classification_report(y_test, y_pred))
    return()
```

# Exploratory Data Analysis

```python
n_bins = 15
limit = 0.85 #threshold to discard variables 
drop_num_variables = [] #list variables that will be dropped 

fig = plt.figure(figsize=(30,25), dpi = 100)
titulo_fig = "numerical_attributes_histogram"

size = int(len(train_num.iloc[0,:]))
n_col = int(5)
n_lines = int(np.ceil(size/n_col))



for i in range(0,size):
    ax = fig.add_subplot(n_lines, n_col, i+1)
    data = train_num.iloc[:,i]
    w = np.zeros_like(data) + 1. / data.size
    values, bins1, patch = plt.hist(data,weights=w, bins = n_bins)
    ax.set_xlabel(train_num.columns[i])
    
    values[::-1].sort()
    if (values[0] > limit):
        drop_num_variables.append(train_num.columns[i])

fig.tight_layout(pad=2)
fig.savefig(titulo_fig, dpi=fig.dpi, format = "pdf")
plt.show() 
```

```python
# Bar plots of counts of categorical atributes
limit = 0.85 #threshold to discard variables 
drop_cat_variables = [] #list variables that will be dropped 


fig = plt.figure(figsize=(30,25), dpi = 100)
titulo_fig = "categorical_attributes_barplot"

size = int(len(train_cat.iloc[0,:]))
n_col = int(5)
n_lines = int(np.ceil(size/n_col))

for i in range(0,size):
    ax = fig.add_subplot(n_lines, n_col, i+1)
    #ax.plot(vec_xyz[i,:,0], vec_xyz[i,:,1], vec_xyz[i,:,2])
    train_cat.iloc[:,i].value_counts(normalize=True, dropna = False).plot(kind='bar')
    value = train_cat.iloc[:,i].value_counts(normalize=True, dropna = False)[0]
    if (value > limit):
        drop_cat_variables.append(train_cat.columns[i])
    ax.set_xlabel(train_cat.columns[i])
    

fig.tight_layout(pad=2)
fig.savefig(titulo_fig, dpi=fig.dpi, format = "pdf")
plt.show() 
```

# Deal with Missing Data

## Numeric Data

- Fractions of missing data

```python
list1 = train_num.isna().sum()/len(train_num.iloc[:,0])
list1
```

- Drop columns if missing data in them is higher than 50%

```python
limit = 0.5
list_drop = []

for i in range(len(list1)):
    if list1[i] > limit:
        list_drop.append(list1.index[i])

if(len(list_drop)>0):
    train_num = train_num.drop(columns = list_drop)
    print("Columns removed:")
    print(list_drop)
```

- List all variables with missing data in train_num

```python
list_col_na = []
list_na = train_num.isna().sum()

for i in range(len(list_na)):
    value = list_na[i]
    if (value > 0):
        list_col_na.append(train_num.columns[i])

print(list_col_na)        
```

- Replace the NA values based on the mean of each column

```python
for col in list_col_na:
    value = train_num[col].median() 
    train_num[col].fillna(value,inplace=True)
```

- Check to see if there is no more NAs

```python
train_num.isna().sum().any()
```

### TEST set
- Can not drop any column here
- Replace missing values with mean value of columns

```python
list_col_na = []
list_na = test_num.isna().sum()

for i in range(len(list_na)):
    value = list_na[i]
    if (value > 0):
        list_col_na.append(test_num.columns[i])

print(list_col_na)  

for col in list_col_na:
    value = test_num[col].median() 
    test_num[col].fillna(value,inplace=True)

       
test_num.isna().sum().any()  
```

## Categorical Data

- Fractions of missing data

```python
list1 = train_cat.isna().sum()/len(train_cat.iloc[:,0])
list1
```

- Drop columns if missing data in them is higher than 50%

```python
limit = 0.5
list_drop = []

for i in range(len(list1)):
    if list1[i] > limit:
        list_drop.append(list1.index[i])

if(len(list_drop)>0):
    train_cat = train_cat.drop(columns = list_drop)
    print("Columns removed:")
    print(list_drop)
```

- List all variables with missing data in train_cat

```python
list_col_na = []
list_na = train_cat.isna().sum()

for i in range(len(list_na)):
    value = list_na[i]
    if (value > 0):
        list_col_na.append(train_cat.columns[i])

print(list_col_na)        
```

- Replace the NA values based on the most common value of each column

```python
for col in list_col_na:
    value = train_cat[col].value_counts().index[0] 
    train_cat.loc[train_cat[col].isna(),col] = value
```

- Check to see if there is no more NAs

```python
train_num.isna().sum().any() 
```

### TEST Data

- List all variables with missing data in test_cat

```python
list_col_na = []
list_na = test_cat.isna().sum()

for i in range(len(list_na)):
    value = list_na[i]
    if (value > 0):
        list_col_na.append(test_cat.columns[i])

print(list_col_na)      

for col in list_col_na:
    value = test_cat[col].value_counts().index[0] 
    test_cat.loc[test_cat[col].isna(),col] = value
    
test_num.isna().sum().any() 
```

# OneHotEncoder for Categorical Data

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

## Train_cat

```python
encoder_train = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(sparse=False),train_cat.columns)], 
                                  remainder='passthrough', verbose_feature_names_out = False)
train_cat_encoded = encoder_train.fit_transform(train_cat)

train_cat_encoded = pd.DataFrame(train_cat_encoded, columns = encoder_train.get_feature_names_out())

train_cat_encoded.head()
```

## Test_cat

```python
encoder_test = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(sparse=False),test_cat.columns)], 
                                  remainder='passthrough', verbose_feature_names_out = False)
test_cat_encoded = encoder_test.fit_transform(test_cat)

test_cat_encoded = pd.DataFrame(test_cat_encoded, columns = encoder_test.get_feature_names_out())

test_cat_encoded.head()
```

# Concatenating Cat e Num

```python
train_final = pd.concat([train_num, train_cat_encoded], axis = 1)
train_final.isna().sum().any()
```

```python
test_final = pd.concat([test_num, test_cat_encoded], axis = 1)
test_final.isna().sum().any()
```

# Separating the target from the TRAIN df

```python
x = train_final.drop('SalePrice', axis = 1)
y = train_final['SalePrice']
```

```python
x.shape
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
lm_y_pred = lm.predict(x_test)
#print_metrics('Linear Regression', y_test,lm_y_pred)

rmse = np.sqrt(metrics.mean_squared_error(y_test,lm_y_pred))

print ('RMSE', rmse)
print(rmse/train['SalePrice'].describe()['mean'])
```
